```python
num_of_tweets = 200
tweets = []
time = []
likes = []

users = '@dragflickscores OR @thedragflick OR @WorldHockeyNews OR @EnglandHockey OR @PolandHockey'
cursor = tweepy.Cursor(api.search_tweets, q = users, tweet_mode = "extended").items(num_of_tweets)

for t in cursor:
    tweets.append(t.full_text)
    time.append(t.created_at)
    likes.append(t.favorite_count)

df = pd.DataFrame({'tweets':tweets, 'likes':likes, 'time':time})
df
```

```python
df = df[~df.tweets.str.contains("RT")]
df = df.reset_index(drop=True)
df
```
